- Primeiramente temos a função de conversão de decimal para binario

```python
arraybits = []
def decimal2Binary(number):

    if number>=1:
        decimal2Binary(number//2)
    arraybits.append(number%2)


decimal2Binary(6)
print(arraybits)
```

- Agora teremos que fazer o complemento a 1

```python
def complementOne(arr):
    for i in range(len(arr)):
        if(arr[i]==0):
            arr[i]=1
        elif arr[i]==1:
            arr[i]=0
            
complementOne(arraybits)
print(arraybits)
```

```python

def complementTwo(arr):
    i=len(arr)-1
    if arr[i]==0:
        arr[i]+=1
    elif arr[i]==1:
        i-=1
        while not (arr[i]==0) and i!=0:
            arr[i]=0
            i-=1
        if arr[i]==1:
            arr[i]=10
        elif arr[i]==0:
             arr[i]=1

complementTwo(arraybits)
print(arraybits)
```

- Agora convertemos o numero para decimal novamente 

```python
def binary2Decimal(arr):
    soma=0
    exp= len(arr) - 1
    for i in range(len(arr)):
        soma+= arr[i]*(2**exp)
        exp -= 1
    return soma

print(binary2Decimal(arraybits))
```
